```python
df = pd.read_excel("Inmuebles.xlsx", header=1)
df = df.drop(columns='Referencia', index=1)
df
```

```python
filtroB = (df["Fecha_Alta"] == pd.to_datetime("29/04/2007")) | (df["Fecha_Alta"] == pd.to_datetime("02/01/2004"))
df[filtroB]
```

```python
filtroC = (df['Tipo'] == "Parking" ) | (df['Tipo'] == "Local" ) | (df['Tipo'] == "Piso" )
df[filtroC]

```
