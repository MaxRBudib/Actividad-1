```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Ventas_totales.csv")
df
```

```python
#Aquí identificamos los valores nulos por columna 
valores_nulos = df.isnull().sum()
valores_nulos
```

```python
#Primero eliminamos columnas innecesarias
df = df.drop(['otros'], axis=1)
df.head()
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
nullVal = df.isnull().sum()
nullVal
```

```python
#Este filtro reemplaza todos los valores nulos de la columna subtotal_ventas_alimentos_bebidas
df2 = df.copy()
df2["subtotal_ventas_alimentos_bebidas"].fillna(round(df["subtotal_ventas_alimentos_bebidas"].mean(),1), inplace=True)
df2.head()
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios

valoresNulos = df2.isnull().sum()
valoresNulos
```

```python
#Este filtro reemplaza los valores nulos por el valor anterior en el registro en las siguientes 3 columnas
df2["salon_ventas"].fillna(method="bfill", inplace=True)
df2["tarjetas_debito"].fillna(method="bfill", inplace=True)
df2["tarjetas_credito"].fillna(method="bfill", inplace=True)
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = df2.isnull().sum()
valoresNulos
```

```python
#En este filtro rellenamos los lugares donde se encuentran valores fijos por los valores maximos y minimos de cada columna
df2["bebidas"].fillna(round(df2["bebidas"].max(),1), inplace=True)
df2["panaderia"].fillna(round(df2["panaderia"].min(),1), inplace=True)
df2["lacteos"].fillna(round(df2["lacteos"].max(),1), inplace=True)
df2["carnes"].fillna(round(df2["carnes"].min(),1), inplace=True)
df2["verduleria_fruteria"].fillna(round(df2["verduleria_fruteria"].max(),1), inplace=True)
df2["alimentos_preparados_rotiseria"].fillna(round(df2["alimentos_preparados_rotiseria"].min(),1), inplace=True)
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = df2.isnull().sum()
valoresNulos
```

```python
#En este filtro modificamos los valores nulos por sus valores siguientes
df2["otros_medios"].fillna(method="ffill", inplace=True)
df2["almacen"].fillna(method="ffill", inplace=True)
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = df2.isnull().sum()
valoresNulos
```

```python
#En este filtro llenamos los valores nulos con la media y la mediana
df2["indumentaria_calzado_textiles_hogar"].fillna(round(df["indumentaria_calzado_textiles_hogar"].mean(),1), inplace=True)
df2["electronicos_articulos_hogar"].fillna(round(df["electronicos_articulos_hogar"].median(),1), inplace=True)
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = df2.isnull().sum()
valoresNulos
```

```python
#Identificar valores nulos por columna para revisar si exitieron cambios
valoresNulos = df2.isnull().sum().sum()
valoresNulos
```

```python
#Por ultimo convertimos el dataframe a csv
df2.to_csv("No_null.csv")
```
