```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Actividad 2.1/No_null.csv')
display(data)
```

```python
valoresNulos = data.isnull().sum()
print(valoresNulos)

valoresNulosCSV = data.isnull().sum().sum()
print(f'\nValores nulos en el CSV: {valoresNulosCSV}')
```

```python
# Primero, generamos una copia de data
data1 = data.copy()

fig = plt.figure(figsize = (10,10))
plt.hist(x=data1['ventas_precios_corrientes'], color='blue', rwidth=0.5)
plt.title('Histograma de Ventas Precios Corrientes')
plt.xlabel('Ventas Precios Corrientes')
plt.ylabel('Frecuencia')

plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
plt.boxplot(data1['ventas_precios_corrientes'])
plt.title('Outliers de Ventas Precios Corrientes')
plt.show()
```

```python
# Encontramos los valores atípicos
y = data1['ventas_precios_corrientes']
limiteSuperior = y.mean() + 3*y.std()
limiteInferior = y.mean() - 3*y.std()

print(f"Limite superior permitido : {round(limiteSuperior,2)}")
print(f"Limite inferior permitido : {round(limiteInferior,2)}")

# Vemos que por el metodo std no se borra ningun valor, por lo tanto, el valor que vemos en el boxplot debe corresponder al metodo iqr
outliers = data1[(y>limiteSuperior) | (y<limiteInferior)]
print(f"\nOutliers:\n{outliers}")
```

```python
data_clean = data1[(y<=limiteSuperior) & (y>=limiteInferior)]
display(data_clean)
```

```python
# Vizualizamos el outlier
fig = plt.figure(figsize = (10,6))
plt.boxplot(data_clean['ventas_precios_corrientes'])
plt.title('Outliers de Ventas Precios Corrientes')
plt.show()
```

```python
y = data1["ventas_precios_corrientes"]

percentile25 = y.quantile(0.25) 
percentile75 = y.quantile(0.75) 

iqr = percentile75 - percentile25

limiteSuperiorIQR = percentile75 + 1.5 * iqr
limiteInferiorIQR = percentile25 - 1.5 * iqr

print(f"Limite superior permitido : {round(limiteSuperiorIQR,2)}")
print(f"Limite inferior permitido : {round(limiteInferiorIQR,2)}")

# La funcion retorna el valor atipico que buscamos
outliers = data1[(y>limiteSuperiorIQR) | (y<limiteInferiorIQR)]
print(f"\nOutliers:\n{outliers}")
```

```python
data_clean_iqr = data1[(y<=limiteSuperiorIQR) & (y>=limiteInferiorIQR)]
display(data_clean_iqr)
```

```python
# Los 4 outliers identificados en un principio ahora ya no son parte del df
fig = plt.figure(figsize = (10,6))
plt.boxplot(data_clean_iqr['ventas_precios_corrientes'])
plt.title('Outliers de Ventas Precios Corrientes')
plt.show()
```

- 3.8 Guardamos el df sin valores atipicos

```python
data_clean_iqr.to_csv("ventas_precios_normales.csv")
```
